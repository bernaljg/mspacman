```python
def normalize_image(char, mean, std):
    char = char.astype("float32")
    for i in range(3):
        m = mean[i]
        s = std[i]
        char[:,:,i] = char[:,:,i] - m
        char[:,:,i] = char[:,:,i]/s
        
    return char
```

```python
def get_avg(char, mean, std):
    char = normalize_image(char, mean, std)
    f_char = np.fliplr(char)
    return (char+f_char)/2
```
